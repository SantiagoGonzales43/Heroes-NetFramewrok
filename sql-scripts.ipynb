# [dbo].[UpdateHeroe]

```sql
/****** Object:  StoredProcedure [dbo].[UpdateHeroe]    Script Date: 24/6/2025 10:41:47 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE PROCEDURE [dbo].[UpdateHeroe]
	@Id INT,
	@Nombre VARCHAR(50),
	@Clase VARCHAR(50),
	@Nivel INT
	
AS
BEGIN
	
	SET NOCOUNT ON;

	UPDATE Heroes 
	SET Nombre = @Nombre, Clase = @Clase, Nivel = @Nivel
	WHERE Id = @Id

	SELECT @@ROWCOUNT AS RowsAffected;
END
GO
```
